```python
# Problem 1: Post Engagement Analyzer

def analyze_engagement(likes_list):
    total_likes = 0

    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like

    # Determine post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Example usage
likes = [200, 300, 250, 300]
analyze_engagement(likes)
```

```python
# Problem 2: Medicine Stock Alert

def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

# Example usage
check_medicine_stock(6)
```

```python
# Problem 3: Rainfall Adequacy Checker

def rainfall_analysis(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

# Example usage
rainfall = [70, 75, 68, 80, 67]
rainfall_analysis(rainfall, 70)
```

```python
# Problem 4: Duplicate Account Detection

def check_duplicate_accounts(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# Example usage
users = ["user1", "user2", "user3", "user1"]
check_duplicate_accounts(users)
```

```python
# Problem 5: Appointment Eligibility

def check_appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

# Example usage
check_appointment_eligibility(21)
```

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)

# Example usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
# Problem 7: Application Health Checker

def check_system_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)

# Example usage
check_system_health(7)
```

```python
# Problem 8: Transaction Limit Checker

def check_transaction(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

# Example usage
check_transaction(60000)
```

```python
# Problem 9: Attendance Eligibility Checker

def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:   # 1 means present
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

# Example usage
attendance = [1, 1, 0, 1, 1]
check_attendance(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0

    # Billing slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)

# Example usage
calculate_electricity_bill(250)
```
