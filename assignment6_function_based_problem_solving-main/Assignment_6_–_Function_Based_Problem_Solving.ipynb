```python
def smart_parking_lot(capacity, logs):
    """
    Manages parking lot entry and exit.
    Returns currently parked vehicles and parking status.
    """
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # Track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full - Capacity Exceeded!")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)
```

```python
smart_parking_lot(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
def estimate_delivery_time(distance, traffic, weather):
    """
    Estimates delivery time based on distance, traffic, and weather.
    """

    # Base time: 5 minutes per km
    eta = distance * 5

    # Traffic impact
    if traffic == "High":
        eta += 15
    elif traffic == "Medium":
        eta += 8

    # Weather impact
    if weather == "Rainy":
        eta += 10
    elif weather == "Storm":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")
```

```python
estimate_delivery_time(8, "High", "Rainy")
```

```python
def seat_occupancy(total_seats, booked_seats):
    """
    Calculates occupancy percentage and show status.
    """

    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
```

```python
seat_occupancy(200, [1]*150)
```

```python
def server_load(cpu_readings):
    """
    Classifies server load based on average CPU usage.
    """

    average = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
```

```python
server_load([45, 60, 70, 85, 90])
```

```python
def classroom_resource_monitor(resources):
    """
    Identifies overused classroom resources.
    Overuse threshold: more than 8 hours
    """

    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
```

```python
classroom_resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
def event_registration(capacity, registrations):
    """
    Manages event registrations and prevents overbooking.
    """

    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
    else:
        waitlist = registrations - capacity
        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", waitlist)
        print("Registration Status: Closed")
```

```python
event_registration(100, 105)
```
